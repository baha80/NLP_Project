```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
print('Libraries loaded ✅')
```

```python
# Load dataset
df = pd.read_csv('../data/sample_reviews.csv')
df['date'] = pd.to_datetime(df['date'])

print(f'Shape: {df.shape}')
print(f'Columns: {df.columns.tolist()}')
df.head()
```

```python
# Basic statistics
print('=== Dataset Info ===')
print(df.info())
print('\n=== Missing Values ===')
print(df.isnull().sum())
print('\n=== Rating Distribution ===')
print(df['rating'].value_counts())
```

```python
# Text length analysis
df['text_length'] = df['text'].apply(len)
df['word_count']  = df['text'].apply(lambda x: len(x.split()))

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].hist(df['text_length'], bins=30, color='#6366f1', edgecolor='white')
axes[0].set_title('Review Length Distribution (chars)')
axes[0].set_xlabel('Character count')

axes[1].hist(df['word_count'], bins=30, color='#06b6d4', edgecolor='white')
axes[1].set_title('Review Length Distribution (words)')
axes[1].set_xlabel('Word count')

plt.tight_layout()
plt.savefig('../assets/eda_length.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'\nAvg chars: {df["text_length"].mean():.0f}')
print(f'Avg words: {df["word_count"].mean():.0f}')
```

```python
# Rating distribution
fig = px.histogram(df, x='rating', nbins=5,
                   color_discrete_sequence=['#6366f1'],
                   title='Rating Distribution')
fig.show()
```

```python
# Platform breakdown
platform_counts = df['platform'].value_counts()
fig = px.pie(values=platform_counts.values,
             names=platform_counts.index,
             title='Reviews by Platform',
             color_discrete_sequence=['#6366f1','#06b6d4','#10b981'])
fig.show()
```

```python
# Word frequency
import nltk
nltk.download('stopwords', quiet=True)
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

all_words = ' '.join(df['text'].tolist()).lower().split()
filtered  = [w for w in all_words if w.isalpha() and w not in stop_words and len(w) > 2]
top_50    = Counter(filtered).most_common(20)

words, counts = zip(*top_50)
fig = px.bar(x=list(counts), y=list(words), orientation='h',
             title='Top 20 Most Frequent Words',
             color=list(counts), color_continuous_scale='viridis')
fig.update_layout(yaxis={'categoryorder': 'total ascending'})
fig.show()
```

```python
# Rating vs text length
fig = px.box(df, x='rating', y='word_count', color='platform',
             title='Word Count by Rating and Platform')
fig.show()
```
