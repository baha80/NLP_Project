# 02 - Preprocessing\nClean reviews, remove duplicates, tokenize text, and build TF-IDF/embedding features.

```python
import sys\nfrom pathlib import Path\n\nROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()\nsys.path.insert(0, str(ROOT))\n\nimport pandas as pd\nfrom preprocessing import TextCleaner, TextVectorizer
```

```python
df = pd.read_csv(ROOT / 'data' / 'reviews.csv')\ndf.head()
```

```python
cleaner = TextCleaner()\nprepared_df, report = cleaner.prepare_dataframe(df, text_col='text', drop_duplicates=True)\nreport
```
